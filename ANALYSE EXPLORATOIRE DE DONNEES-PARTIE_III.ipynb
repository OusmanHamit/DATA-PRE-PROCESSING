```python
def lowerCase(df):
    df.columns = [colname.lower() for colname in df.columns]
```

```python
lowerCase(df)
```

```python
df.head()
```

```python
def upperCase(df):
    df.columns = [colname.upper() for colname in df.columns]
```

```python
upperCase(df)
```

```python
df.head()
```

```python
def lowerCaseContent(df):
    for colname in df.select_dtypes(['object']):        
        df[colname] = df[colname].str.lower()
```

```python
lowerCaseContent(df)
```

```python
df.head()
```

```python
def upperCaseContent(df):
    for colname in df.select_dtypes(['object']):        
        df[colname] = df[colname].str.upper()
```

```python
upperCaseContent(df)
```
