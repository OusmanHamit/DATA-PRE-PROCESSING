```python
import pandas as pd
import numpy as np 
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
%matplotlib notebook 
```

```python
df = pd.read_csv('Salary.csv')
df.head()
```

```python
X = df.iloc[:,0].values
y = df.iloc[:,1].values
```

- ***Diviser le DataSet***

```python
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=.2,
                                                    random_state = 3)
```

```python
#plt.scatter(X_train, y_train, label = 'Donnees d\'entrainement', color = 'r')
#plt.scatter(X_test, y_test, label = 'Donnees d\'evaluation', color = 'b')
#plt.xlabel('Experiences')
#plt.ylabel('Salaires')
#plt.title('Regression lineaire simple')
#plt.legend()
#plt.show();
```

- ***Entrainer le model*** 

```python
reg = LinearRegression()
```

```python
reg.fit(X_train, y_train);
```

- ***Prédiction de salaires***

```python
y_pred = reg.predict(X_test)
```

- ***Comparer les valeurs prédites par notre model aux valeurs réelles***

```python
print('y_test \n', y_test)
print('\n ================================= \n')
print('y_pred \n', y_pred)
```

- ***Répresentation graphique***

```python
plt.plot(X_train, reg.predict(X_train), color = 'b', label='Droite de regression')
plt.scatter(X_test, y_test, color = 'g', label = 'Données réelles du TestSet')
plt.xlabel('Experiences')
plt.ylabel('Salaires')
plt.title('Prédictions de salaires')
plt.legend()
plt.show();
```
