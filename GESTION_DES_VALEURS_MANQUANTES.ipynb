* **Imputation par la valeur la plus frequente**

```python
missing_values = ['/','--','na','n/a'] 
cp = pd.read_csv("data.csv", na_values=missing_values)
```

```python
cp['Age'].fillna(0, inplace=True)
```

```python
cp.shape
```

```python
simpleImputer = SimpleImputer(missing_values=0, strategy='mean')
cp['Age'] = simpleImputer.fit_transform(cp['Age'].values.reshape(22,1))
```

```python
cp['Age']
```

```python
cp['Salaire_Bourse'].fillna(0, inplace=True)
```

* **Imputation par le mode (most frequent)**

```python
simpleImputer = SimpleImputer(missing_values=0, strategy='most_frequent')
cp['Salaire_Bourse'] = simpleImputer.fit_transform(cp['Salaire_Bourse'].values.reshape(22,1))
```
