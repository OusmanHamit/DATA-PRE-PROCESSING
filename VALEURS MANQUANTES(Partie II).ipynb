```python
df['product_type'].value_counts()[0]
```

**Sélectionner les colonees qui nous interesse**

```python
features = ['brand','name','product_type','img','description','shade_img']
df = df[features]
```

```python
df['brand'].value_counts()
```

```python
def handle_categorical_features(feature, df):
        df[feature].fillna(df[feature].value_counts().index[0], inplace = True)
```

```python
for feature in df.columns:
    handle_categorical_features(feature, df)
```

```python
showme_all_mv(df)
```
